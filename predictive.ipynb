```python
df_long=data_1.melt(
    id_vars=["Indicator Name","Country Name"],
    var_name="year",
    value_name="value"
)
df_long=df_long[df_long["year"].str.isnumeric()]
df_long["year"]=df_long["year"].astype(int)
df_final=df_long.pivot_table(
    index="year",
    columns="Indicator Name",
    values="value"
).reset_index()
merge_data["year"]=pd.to_datetime(merge_data["time"],errors="coerce").dt.year

final_data=pd.merge(
    merge_data,
    df_final,
    on="year",
    how="left"
)
```

```python
final_data.shape
```

```python
final_data=final_data.dropna(axis=1,thresh=len(final_data)*0.5)
final_data.isna().sum().sort_values(ascending=False)
```

```python
final_data=final_data.fillna(method="ffill")
```

```python
final_data.nunique().sort_values()
final_data=final_data.loc[:,final_data.nunique()>1]
```

```python
final_data.head()
```

```python
final_data["month"]=final_data["datetime"].dt.month
final_data["day"]=final_data["datetime"].dt.day
final_data["hour"]=final_data["datetime"].dt.hour
```

```python
final_data.head()
```

```python
final_data=final_data.drop(["time","datetime"],axis=1)
```

```python
corr=final_data.corr(numeric_only=True)["demand_mw"].sort_values(ascending=False)
print(corr.head(20))
```

```python
important_cols=corr[abs(corr)>0.3].index.tolist()
data_model=final_data[important_cols]
```

```python
data_model.shape
```

```python
corr_matrix = data_model[important_cols].corr().abs()

upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

to_drop = [col for col in upper.columns if any(upper[col] > 0.9)]

final_features = [col for col in important_cols if col not in to_drop]
df=data_model[final_features]
```

```python
df.shape
```

```python
df.head()
```

```python
df["year"]=pd.to_datetime(merge_data["time"],errors="coerce").dt.year
df["day"]=pd.to_datetime(merge_data["time"],errors="coerce").dt.day
df["hour"]=pd.to_datetime(merge_data["time"],errors="coerce").dt.hour
df["month"]=pd.to_datetime(merge_data["time"],errors="coerce").dt.month
```

```python
df['lag1'] = df['demand_mw'].shift(1)
df['lag24'] = df['demand_mw'].shift(24)
df['rolling_mean_24'] = df['demand_mw'].rolling(24).mean()

df['hour_sin'] = np.sin(2*np.pi*df['hour']/24)
df['hour_cos'] = np.cos(2*np.pi*df['hour']/24)
```

```python
train = df[df["year"] <= 2023]
test = df[df["year"] == 2024]

X_train = train.drop(columns=["demand_mw"])
y_train = train["demand_mw"]

X_test = test.drop(columns=["demand_mw"])
y_test = test["demand_mw"]
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
